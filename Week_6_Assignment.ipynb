```python
# Importing the math module to use mathematical functions like sqrt()
import math

# Global variable accessible by all functions
station_name = "Kathmandu Weather Station"

# List containing daily temperature readings
temperatures = [18.4, 22.1, 15.7, 29.3, 11.8, 25.6, 19.2]


# Function to calculate and return the average temperature
def get_average(temps):
    return sum(temps) / len(temps)


# Function to calculate standard deviation
def get_deviation(temps):

    # Local variable 'mean' only exists inside this function
    mean = get_average(temps)

    # Calculate variance using the average
    variance = 0
    for temp in temps:
        variance += (temp - mean) ** 2

    variance = variance / len(temps)

    # Standard deviation is the square root of variance
    deviation = math.sqrt(variance)

    return deviation


# Function to display the complete temperature summary
def get_summary(temps):

    print("Station:", station_name)
    print("Minimum Temperature:", min(temps), "°C")
    print("Maximum Temperature:", max(temps), "°C")
    print("Average Temperature:", round(get_average(temps), 2), "°C")
    print("Standard Deviation:", round(get_deviation(temps), 2))


# Calling the function to display the summary
get_summary(temperatures)


# The variable 'mean' cannot be accessed here because it is a local variable
# print(mean)
# This would cause a NameError because 'mean' only exists inside get_deviation()
```

```python
#Question 2
# Importing random module to randomly select a friend
import random

# Setting a seed so the random result remains the same every time
random.seed(42)

# List of friends and total restaurant bill
friends = ["Ramesh", "Sunita", "Bikash", "Anjali", "Dipak"]
total_bill = 3750


# Function to divide the bill equally among all friends
def split_bill(friends, total):
    return total / len(friends)


# Function to randomly select the lucky person
def pick_lucky(friends):
    return random.choice(friends)


# Function to display final payment summary
def final_summary(friends, total):

    # Getting the equal share for each friend
    share = split_bill(friends, total)

    # Selecting the lucky friend
    lucky_person = pick_lucky(friends)

    print("Restaurant Bill Summary")
    print("------------------------")

    # Printing each friend's normal share
    for person in friends:
        print(person, "pays NPR", share)

    # Local variable storing the lucky person's final amount
    lucky_total = share + 50

    print("\nLucky Person:", lucky_person)
    print(lucky_person, "pays an extra NPR 50")
    print("Final amount paid by lucky person: NPR", lucky_total)


# Calling the summary function
final_summary(friends, total_bill)
```

```python
#Question 3
# Importing datetime module to work with dates
import datetime


# Global variable for the college name
college_name = "Bhaktapur Multiple Campus"


# Exam starting date and subjects with day offsets
start_date = "2025-05-01"

exams = [
    ("Python Programming", 0),
    ("Data Structures", 3),
    ("Database Systems", 6),
    ("Computer Networks", 10),
    ("Mathematics", 14)
]


# Function converts a date string into a datetime object
def parse_date(date_str):
    return datetime.datetime.strptime(date_str, "%Y-%m-%d")


# Function calculates the exam date after adding certain days
def get_exam_date(start_str, days):

    # Convert string into datetime object
    start = parse_date(start_str)

    # Add days using timedelta
    exam_day = start + datetime.timedelta(days=days)

    # Convert datetime object back into string format
    return exam_day.strftime("%Y-%m-%d")


# Function prints the complete exam schedule
def print_schedule(start_str, exams):

    print(college_name)
    print("Exam Schedule")
    print("---------------------")

    # Loop through each subject and calculate its date
    for subject, days in exams:
        date = get_exam_date(start_str, days)
        print(subject, ":", date)


# Calling the function to print the schedule
print_schedule(start_date, exams)
```

```python
%%writefile discount.py

# Global constant for VAT rate
TAX_RATE = 0.13


# Function to apply discount to the original price
def apply_discount(price, percent):
    discount_amount = price * (percent / 100)
    return price - discount_amount


# Function to add VAT to the discounted price
def apply_tax(price):
    return price + (price * TAX_RATE)


# Function to calculate the final price
# Discount is applied first and then VAT is added
def final_price(price, discount_pct):
    discounted_price = apply_discount(price, discount_pct)
    return apply_tax(discounted_price)
```
