```python
#Question 1 - Cooperative Bank and balance - Bank Account Manager question
# Creating a BankAccount class to manage customer accounts
class BankAccount:

    # constructor to initialize name, account number and balance
    def __init__(self, name, account_number, balance=0):
        self.name = name
        self.account_number = account_number
        self.balance = balance

    # method to deposit money into account
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited into {self.account_number}")

    # method to withdraw money
    def withdraw(self, amount):
        # checking if enough balance is available
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn from {self.account_number}")

    # method to display current balance
    def get_balance(self):
        print(f"{self.name} - Balance: {self.balance}")


# creating accounts from given data
accounts_data = [
    ("Ramesh Thapa", "A001", 5000),
    ("Sunita Karki", "A002", 0),
    ("Bikash Rai", "A003", 12000),
]

# storing objects in a list
accounts = []

for data in accounts_data:
    acc = BankAccount(data[0], data[1], data[2])
    accounts.append(acc)

# performing required transactions
accounts[1].deposit(3000)      # A002 deposit
accounts[2].withdraw(15000)    # A003 withdraw (should fail)
accounts[0].withdraw(2000)     # A001 withdraw

print("\nFinal Balances:")
for acc in accounts:
    acc.get_balance()
```

```python
#Question 2 - Student Report Card- Name and marks in 5 subjects
# Creating Student class to handle marks and results
class Student:

    # storing student name and marks list
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # method to calculate average marks
    def average(self):
        return sum(self.marks) / len(self.marks)

    # method to calculate grade based on average
    def grade(self):
        avg = self.average()

        if avg >= 80:
            return "A"
        elif avg >= 65:
            return "B"
        elif avg >= 50:
            return "C"
        elif avg >= 40:
            return "D"
        else:
            return "F"

    # method to display student result
    def display(self):
        avg = self.average()
        grade = self.grade()

        # checking pass or fail
        result = "Pass" if avg >= 40 else "Fail"

        print(f"\nName: {self.name}")
        print(f"Average: {avg:.2f}")
        print(f"Grade: {grade}")
        print(f"Result: {result}")


# given student data
students_data = [
    ("Aarav",  [78, 85, 60, 90, 72]),
    ("Sita",   [45, 50, 38, 60, 55]),
    ("Bishal", [30, 25, 40, 35, 28]),
    ("Priya",  [90, 88, 95, 92, 87]),
]

# creating objects and displaying results
students = []

for data in students_data:
    s = Student(data[0], data[1])
    students.append(s)

for s in students:
    s.display()
```

```python
#Question 3 - Food Delivery
# Parent class for all delivery partners
class DeliveryPartner:

    # initializing common data
    def __init__(self, name, partner_id, deliveries):
        self.name = name
        self.partner_id = partner_id
        self.deliveries = deliveries

    # method to be overridden in child classes
    def total_earning(self):
        return 0

    # display common info
    def display(self):
        print("\nName:", self.name)
        print("Deliveries:", self.deliveries)
        print("Total Earning:", self.total_earning())


# Bike Rider class
class BikeRider(DeliveryPartner):

    # adding extra attribute km_travelled
    def __init__(self, name, partner_id, deliveries, km_travelled):
        super().__init__(name, partner_id, deliveries)
        self.km_travelled = km_travelled

    # earning calculation for bike rider
    def total_earning(self):
        return (self.deliveries * 80) + (self.km_travelled * 5)


# Walker class
class Walker(DeliveryPartner):

    # adding rainy deliveries
    def __init__(self, name, partner_id, deliveries, rainy_deliveries):
        super().__init__(name, partner_id, deliveries)
        self.rainy_deliveries = rainy_deliveries

    # earning calculation for walker
    def total_earning(self):
        return (self.deliveries * 60) + (self.rainy_deliveries * 50)


# Car Driver class
class CarDriver(DeliveryPartner):

    # adding fuel cost
    def __init__(self, name, partner_id, deliveries, fuel_cost):
        super().__init__(name, partner_id, deliveries)
        self.fuel_cost = fuel_cost

    # earning calculation for car driver
    def total_earning(self):
        return (self.deliveries * 120) - self.fuel_cost


# given partner data
partners = [
    BikeRider("Santosh Rai", "B-01", 15, 42),
    Walker("Kabita Maharjan", "W-01", 18, 5),
    CarDriver("Roshan KC", "C-01", 20, 850),
]

# printing all partner details
highest_earner = None
max_earning = 0

for p in partners:
    p.display()

    # finding highest earner
    earning = p.total_earning()
    if earning > max_earning:
        max_earning = earning
        highest_earner = p

# printing highest earning partner
print("\nHighest Earner:")
print(highest_earner.name, "-", max_earning)
```

```python
#Question 4 - Bus Ticket Booking System
# creating Bus class for ticket booking system
class Bus:

    # initializing bus details
    def __init__(self, route, total_seats):
        self.route = route
        self.total_seats = total_seats
        self.booked = []   # list to store booked seats

    # method to book a seat
    def book_seat(self, seat_number, passenger_name):

        # checking if seat already booked
        for seat in self.booked:
            if seat[0] == seat_number:
                print("Seat already booked")
                return

        # if not booked, add to list
        self.booked.append((seat_number, passenger_name))
        print(f"Seat {seat_number} booked for {passenger_name}")

    # method to count available seats
    def available_seats(self):
        return self.total_seats - len(self.booked)

    # method to show all passengers
    def passenger_list(self):
        print("\nPassenger List:")
        for seat, name in self.booked:
            print(f"Seat {seat} - {name}")


# creating bus object
bus = Bus("Kathmandu - Pokhara", 10)

# booking list
bookings = [
    (3, "Ramila Shrestha"),
    (7, "Deepak Gurung"),
    (3, "Anita Rai"),        # duplicate
    (1, "Prakash Magar"),
    (7, "Suman Tamang"),     # duplicate
]

# processing bookings
for seat, name in bookings:
    bus.book_seat(seat, name)

# showing results
print("\nAvailable Seats:", bus.available_seats())
bus.passenger_list()
```
