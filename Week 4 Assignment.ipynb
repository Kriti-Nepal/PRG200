```python
#Question 1

# Inventory data of the shop
inventory = {
    "rice":  {"price": 120, "stock": 20},
    "milk":  {"price": 90,  "stock": 10},
    "bread": {"price": 60,  "stock": 15},
    "eggs":  {"price": 15,  "stock": 30}
}

# Customer cart
cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}

# Function to process order
def process_order(inventory, cart):

    total = 0  # grand total

    print("---- Bill ----")

    # loop through cart items
    for item in cart:
        qty = cart[item]

        # check item exists in inventory
        if item in inventory:

            # check stock availability
            if inventory[item]["stock"] >= qty:

                # calculate cost
                cost = inventory[item]["price"] * qty

                # update stock
                inventory[item]["stock"] -= qty

                # add to total
                total += cost

                # print item bill
                print(item, "x", qty, "= NPR", cost)

            else:
                print("Sorry, not enough stock for", item)

        else:
            print(item, "not available")

    # print grand total
    print("Grand Total: NPR", total)

    print("-------------- Updated stock:")

    # print updated inventory
    for item in inventory:
        print(item + "=", inventory[item]["stock"])


# Call function
process_order(inventory, cart)
```

```python
#Question 2
# List of water level sensors (location, water level in meters)
sensors = [
    ("Chatara", 2.8),
    ("Tribeni Ghat", 5.4),
    ("Koshi Barrage", 4.1),
    ("Sunsari Bridge", 1.9),
    ("Saptakoshi Camp", 6.0),
]

# Function to check water level status
def check_water_level(location, level_metres):

    # Safe condition
    if level_metres < 3:
        return "Safe"

    # Warning condition
    elif level_metres <= 5:
        return "Warning — Alert nearby villages"

    # Danger condition
    else:
        return "DANGER — Evacuate immediately!"


# Loop through each sensor and print result
for location, level in sensors:
    result = check_water_level(location, level)
    print(location, "(", level, "m):", result)
```

```python
#Question 3
# List of BS months
bs_months = [
    "Baisakh","Jestha","Ashadh","Shrawan","Bhadra","Ashwin",
    "Kartik","Mangsir","Poush","Magh","Falgun","Chaitra"
]

# Function to convert date between BS and AD
def convert_date(date_str, from_cal, to_cal):

    # If same calendar, return unchanged
    if from_cal == to_cal:
        return date_str

    # split date into year, month, day
    year, month, day = map(int, date_str.split("-"))

    # AD to BS conversion
    if from_cal == "AD" and to_cal == "BS":
        year += 56  # add 56 years
        return f"{day} {bs_months[month-1]}, {year} BS"

    # BS to AD conversion
    elif from_cal == "BS" and to_cal == "AD":
        year -= 56  # subtract 56 years
        return f"{year}-{month:02d}-{day:02d} AD"


# Customer data
customers = [
    {"name": "Kriti Nepal", "date": "2063-01-26", "cal": "BS", "need": "AD", "style": "full"},
    {"name": "Ramesh Thapa", "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki", "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai", "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"},
]

# Loop through customers and convert dates
for c in customers:

    # convert date using function
    converted = convert_date(c["date"], c["cal"], c["need"])

    # print result
    print(c["name"], "| Original:", c["date"], c["cal"], "| Converted:", converted)
```

```python
#Question 4
# Function to find word frequency
def word_frequency(text):

    # convert text to lowercase so words match correctly
    text = text.lower()

    # remove punctuation (.,!?)
    for ch in ".,!?":
        text = text.replace(ch, "")

    # split text into words
    words = text.split()

    # dictionary to store word counts
    freq = {}

    # count each word
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

    # sort words by frequency (highest first)
    sorted_words = sorted(freq.items(), key=lambda x: x[1], reverse=True)

    # print top 3 words
    print("Top 3 words:")
    for i in range(3):
        print(sorted_words[i][0], "—", sorted_words[i][1], "times")


# test text
text = """ Nepal is a beautiful country. Nepal has Mount Everest. Everest is the highest mountain in the world. Many tourists visit Nepal every year to see Everest and other mountains. Nepal is known for its mountains and natural beauty. """

# call function
word_frequency(text)
```

```python
#Question 5
# Account data
accounts = {
    "A001": {"name": "Ramesh Thapa", "balance": 15000, "pin": "1234"},
    "A002": {"name": "Sunita Karki", "balance": 8500, "pin": "5678"},
    "A003": {"name": "Bikash Rai", "balance": 22000, "pin": "9012"}
}

# ATM function
def atm(account_id, pin, action, amount=0):

    # check if account exists
    if account_id not in accounts:
        print("Account not found")
        return

    # get account details
    acc = accounts[account_id]

    # check PIN
    if acc["pin"] != pin:
        print("Incorrect PIN")
        return

    # check balance
    if action == "balance":
        print("Name:", acc["name"])
        print("Balance:", acc["balance"])

    # deposit money
    elif action == "deposit":
        acc["balance"] += amount
        print("Deposit successful")
        print("New Balance:", acc["balance"])

    # withdraw money
    elif action == "withdraw":

        # check enough balance
        if acc["balance"] >= amount:
            acc["balance"] -= amount
            print("Withdrawal successful")
            print("Remaining Balance:", acc["balance"])

        else:
            print("Insufficient funds")


# Test calls
atm("A001", "1234", "balance")

atm("A002", "0000", "withdraw", 2000)   # wrong PIN

atm("A002", "5678", "deposit", 3000)

atm("A003", "9012", "withdraw", 25000)  # insufficient funds

atm("A004", "1111", "balance")          # account not found
```
